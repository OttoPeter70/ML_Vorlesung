```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IceCream.csv')
```

```python
plt.scatter(df['Temperature'], df['SoldIceCream'])
plt.xlabel("Temperature")
plt.ylabel("SoldIceCream")
```

```python
def train(X, Y, steps, eta):
    w_0 = w_1 = 0
    for i in range(steps):
        grad_w_0 = grad_w_1 = 0
        for j in range(len(X)):
            grad_w_0 = grad_w_0 - Y[j]+ w_1*X[j] + w_0
            grad_w_1 = grad_w_1 - X[j]*Y[j]+ w_1*X[j]*X[j]+w_0 * X[j]
        w_0 = w_0 - eta * grad_w_0
        w_1 = w_1 - eta * grad_w_1
    return [w_0, w_1]
```

```python
a = train(df['Temperature'], df['SoldIceCream'], 200000, 0.0001)
```

```python
a
```

```python
def predict(x, w_0, w_1):
    return w_1*x+w_0
```

```python
xmax = df['Temperature'].max()
xmin = df['Temperature'].min()
ymax = predict(xmax, a[0], a[1])
ymin = predict(xmin, a[0], a[1])
```

```python
print(ymax)
print(ymin)
```

```python
ax = plt.axes()
ax.scatter(df['Temperature'], df['SoldIceCream'])
ax.plot([xmin, xmax],[ymin, ymax])
plt.xlabel("Temperature")
plt.ylabel("SoldIceCream")
plt.show()
```
